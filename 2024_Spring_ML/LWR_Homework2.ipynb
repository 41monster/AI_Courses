```python
# Necessary imports
# Practice: Find the reason why the accuracy of locally weighted linear regression is not satisfactory even as tau increases.
# Homework: Please modify the code to calculate parameters by gradient descent rather than normal equation.
import numpy as np
from ipywidgets import interact
from bokeh.plotting import figure, show, output_notebook
from bokeh.layouts import gridplot
import matplotlib.pyplot as plt
%matplotlib inline
output_notebook()
plt.style.use('seaborn-dark')

# function to perform locally weighted linear regression
def local_weighted_regression(x0, X, Y, tau):
    # add bias term
    x0 = np.r_[1, x0]
    X = np.c_[np.ones(len(X)), X]

    # fit model: normal equations with kernel
    xw = X.T * weights_calculate(x0, X, tau)
    theta = np.linalg.pinv(xw @ X) @ xw @ Y
    # "@" is used to
    # predict value
    return x0 @ theta

# function to perform weight calculation
def weights_calculate(x0, X, tau):
    return np.exp(np.sum((X - x0) ** 2, axis=1) / (-2 * (tau **2) ))

# plot locally weighted regression for different bandwidth values
def plot_lwr(tau):
    # prediction
    domain = np.linspace(-3, 3, num=5)
    prediction = [local_weighted_regression(x0, X, Y, tau) for x0 in domain]

    plot = figure()
    plot.title.text = 'tau=%g' % tau
    plot.scatter(X, Y, alpha=.3)
    plot.line(domain, prediction, line_width=2, color='red')

    return plot

#define distribution
n = 1000

# generate dataset
X = np.linspace(-3, 3, num=n)
Y = np.abs(X ** 3 - 1)

# jitter X
X += np.random.normal(scale=.1, size=n)

# show the plots for different values of Tau
show(gridplot([
	[plot_lwr(10.), plot_lwr(1.)],
	[plot_lwr(0.1), plot_lwr(0.01)]
]))
```
